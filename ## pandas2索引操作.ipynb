```python
import pandas as pd
```

```python
df = pd.read_csv('./data/train.csv')
```

```python
df['Age'][:6]
```

# 上面就是索引一个列的操作

```python
df[['Age','Fare']][:5]
```

# 跟上面的不太一样，必须是：df[['索引1'，‘索引2’]][m:n]
# 上面注意有两层方括号

```python
# df[0]      # 此时会报错，
```

```python
df[:5]
```

# loc：用label去定位
# iloc: 用position去定位

```python
df.iloc[0]
```

# 总结一下，不能直接用df[n]进行索引，只能用：
## df.loc[n]
## df.iloc[m]   查看第m行数据
## df.iloc[m:n]   查看第m到第n行数据   (中间用冒号分开)

```python
df.iloc[1]
```

```python
df.iloc[2:5]
```

# -------------------------------------------------------------------------

# 如果不想把所有列都拿出来，可以进行筛选
# 格式为：df.iloc[m:n,a:b]
# 上面的意思是：选择m到n行，a到b列

```python
df.iloc[1:8,0:5]
```

# --------------------------------------------------------------------

# 下面来看loc的用法

```python
df = df.set_index('Name')
```

```python
# df['Allen, Mr. William Henry']   # 这样会报错
```

```python
df.loc['Allen, Mr. William Henry']
```

# 说明直接用df['名字'] 是不对的
# df.loc['名字']

# iloc与loc的区别：
# iloc 是用值进行索引
# loc 是用标签进行索引

```python
df.loc['Allen, Mr. William Henry','Age']
```

```python
df.loc['Allen, Mr. William Henry','Fare']
```

# 上面的例子很重要

```python
df.loc['Cumings, Mrs. John Bradley (Florence Briggs Thayer)':'Palsson, Master. Gosta Leonard']
```

# df.loc['A':'B']

# ---------------------------------------------------------------------------------

# 另外，当索引到某个元素时可以直接进行赋值操作

```python
df.loc['Cumings, Mrs. John Bradley (Florence Briggs Thayer)','Age']=100000
```

```python
df
```

# bool 类型索引

```python
df['Fare']>40
```

# 这样就返回一个bool类型索引

```python
df[df['Fare']>60][:6]
```

# 上面的例子说明bool类型可以传回dateframe当中来,进行筛选

```python
df.head()
```

```python
df[df['Sex']=='male'][:5]
```

# df[df['Sex']=='male'][:5]    将bool类型返回dataframe中，从而筛选出满足条件的结果

# 对上面满足条件的结果进行统计计算：

```python
df.loc[df['Sex']=='male','Age'].mean()
```

```python
(df['Age'] > 70).sum()
```
