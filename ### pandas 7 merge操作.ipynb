```python
import pandas as pd
```

```python
left=pd.DataFrame({'key':['k0','k1','k2','k3'],
                     'A':['A0','A1','A2','A3'],
                      'B':['B0','B1','B2','B3']})
```

```python
right=pd.DataFrame({'key':['k0','k1','k2','k3'],
                     'C':['C0','C1','C2','C3'],
                      'D':['D0','D1','D2','D3']})
```

```python
left
```

```python
right
```

# 上面用字典创建了DataFrame结构

```python
pd.merge(left,right)
```

# pd.merge(left,right)     使用merge将左表和右表合并

```python
pd.merge(left,right,on='key')
```

# pd.merge(left,right,on='key')  通过on 指定重复的地方

```python
left=pd.DataFrame({'key1':['k0','k1','k2','k3'],
                   'key2':['k0','k1','k2','k3'],
                     'A':['A0','A1','A2','A3'],
                      'B':['B0','B1','B2','B3']})
```

```python
right=pd.DataFrame({'key1':['k0','k1','k2','k3'],
                    'key2':['k0','k1','k2','k3'],
                     'C':['C0','C1','C2','C3'],
                      'D':['D0','D1','D2','D3']})
```

```python
left
```

```python
right
```

```python
pd.merge(left,right,on=['key1','key2'])
```

#  ------------------------------------------------------------------------------

```python
left=pd.DataFrame({'key1':['k0','k1','k2','k3'],
                   'key2':['k0','k1','k2','k4'],
                     'A':['A0','A1','A2','A3'],
                      'B':['B0','B1','B2','B3']})
right=pd.DataFrame({'key1':['k0','k1','k2','k3'],
                    'key2':['k0','k1','k2','k3'],
                     'C':['C0','C1','C2','C3'],
                      'D':['D0','D1','D2','D3']})
```

# 上面构造的数据两个key2不完全一样

```python
pd.merge(left,right,on=['key1','key2'])
```

# 直接删除了元素不同的行，然后进行合并

```python
pd.merge(left,right,on=['key1','key2'],how='outer')
```

# 相当于求并集

```python
pd.merge(left,right,on=['key1','key2'],how='outer',indicator=True)
```

```python
pd.merge(left,right,on=['key1','key2'],how='left',indicator=True)    
```

# 上面指定了以左表为基准

```python
pd.merge(left,right,on=['key1','key2'],how='left')    
```
